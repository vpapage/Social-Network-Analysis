####<b>Σχολιάστε εδώ τα αποτελέσματα των ερωτημάτων Β1, Β2, Β3.</b>

##### Τοπολογία REG <b>(Regular)</b>
Κύριο χαρακτηριστικό του κανονικού γράφου είναι πως όλες οι κορυφές του έχουν ίδιο βαθμό κόμβου, με μηδενική τυχαιότητα (p=0).
Έτσι τα αποτελέσματα που παίρνουμε είναι λογικά. Ο μέσος βαθμός κόμβου  είναι 4 γιατί ορίσαμε $d=4$ και η διασπορά των βαθμών κόμβου είναι μηδενική. 
##### Τοπολογία RGER (<b>Random Graph Erdos-Renyi</b>)
Στο μοντέλο  RGER(n,M),ένα γράφημα επιλέγεται ομοιόμορφα τυχαία από τη συλλογή όλων των γραφημάτων που έχουν κόμβους n και ακμές Μ. Ο βαθμός των κόμβων ακολουθεί κατανομή Poisson και αυτό το παρατηρούμε στην γραφική παράσταση <b>"Distribution of Node Degree RGER Graph"</b>. Ενώ περιμένουμε η μέση τιμή και η διασπορά του βαθμού κόμβων να είναι πολύ κοντά, στην δικιά μας περίπτωση είναι σχεδόν ίσες. 
##### Τοπολογία RRG (<b>Random Geometric Graph</b>) 
Ένα μαθηματικά απλό χωρικό δίκτυο, ένα μη κατευθυνόμενο γράφημα RGG(n,r,l) που κατασκευάζεται τοποθετώντας τυχαία N κόμβους σε ένα χώρο $lxl$ (σύμφωνα με μια καθορισμένη κατανομή πιθανότητας) και τους συνδέοντάς του ανν απέχουν λιγότερο από την ακτίνα r που έχει δοθεί σαν όρισμα. Ο βαθμός των κόμβων ακολουθεί ομοιόμορφη κατανομή και αυτό το παρατηρούμε στην γραφική παράσταση. Η μέση τιμή και η διασπορά του βαθμού κόμβων έχουν μεγάλη απόκλιση. 
##### Τοπολογία SF (<b>Scale-Free Network</b>) 
Ένα δίκτυο του οποίου η κατανομή βαθμoυ κόμβων ακολουθεί τον εκθετικό νόμο. Μικρό ποσοστό των κόμβων έχει πολύ μεγάλο βαθμό κόμβου (τα λεγόμενα HUB) με τους υπόλοιπους κόμβους να σχηματίζουν πολύ λίγους συνδέσμους.
Αυτό σημαίνει πως και η διασπορά του βαθμού κόμβων θα είναι αυξημένη. Όλα αυτά επαληθεύονται κι από τα πειραματικά μας δεδομένα.

##### Τοπολογία SW (<b>Small-World Network</b>) 
Oι περισσότεροι κόμβοι δεν είναι γείτονες μεταξύ τους, αλλά οι γείτονες οποιουδήποτε δεδομένου κόμβου είναι πιθανό εκείνοι να είναι γείτονες μεταξύ τους. Οι περισσότεροι κόμβοι μπορούν να προσεγγιστούν από κάθε άλλο κόμβο με έναν μικρό αριθμό βημάτων (hops). Συγκεκριμένα, ένα δίκτυο small-world ορίζεται ως ένα δίκτυο όπου η τυπική απόσταση L μεταξύ δύο τυχαία επιλεγμένων κόμβων (ο αριθμός των hops) αυξάνεται λογαριθμικά με τον αριθμού των κόμβων N στο δίκτυο, δηλαδή: $L ->logN$ 
Υπάρχει λίγοι κόμβοι στο δίκτυο με μεγάλο αριθμό συνδέσεων (υψηλό βαθμό κόμβου). Αυτοί οι κόμβοι χρησιμεύουν ως οι κοινές συνδέσεις που μεσολαβούν στα μικρά μήκη διαδρομής μεταξύ άλλων άκρων.   

### Γ) Δίκτυα με βάρη

Για ένα δίκτυο με βάρη στις συνδέσεις του, όμοια με τον πίνακα γειτνίασης, ορίζεται ο πίνακας βαρών, $W=[w_{ij}]$, όπου με $w_{ij}>0$ συμβολίζεται το βάρος στην ακμή που ενώνει τους κόμβους $i,j$. Για μη-κατευθυνόμενους γράφους ο πίνακας $W$ είναι συμμετρικός.

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να δημιουργηθεί τυχαίος πίνακας βαρών. Το εύρος των βαρών να κυμαίνεται στο διάστημα $[1,10]$.

```python
# για τιμές στο [1,10]: α=1, b=10, weight=np.random.rand()*(b-a) + a
a=1
b=10

def array_line(n):
    return [np.random.rand()*(b-a) + a for i in range(n)]

def adjacent_array(n):
    # Adjacent Weighted Array
    weight_array = np.array([array_line(n) for i in range(n)])
    # Symmetric Array
    symmetric = (np.maximum(weight_array, weight_array.transpose()))
    # (i,i) = 0
    for i in range(n):
        symmetric[i][i] = 0
    return symmetric
    
# Adding weighted edge
def add_weight(G, n):
    adj = adjacent_array(n)
    
    for e in G.edges():
        n0 = e[0]
        n1 = e[1]
        G[n0][n1]['weight'] = adj[n0][n1]
    #print(G.adj)
    return G

REG_weighted= add_weight(REG,n)
RGER_weighted= add_weight(RGER,n)
RGG_weighted= add_weight(RGG,n)
SF_weighted= add_weight(SF,n)
SW_weighted= add_weight(SW,n)
```

##### i) Να υπολογιστεί η δύναμη του κάθε κόμβου

```python
#δίνεται η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α:
def strength(T):
    weights_list=[]
    for i in range(0,len(T.adj)):      #for each node 
      a=[]
      for j in range (0,len(T.adj[i].items())):
        d = list (T.adj[i].values())[j]
        c = list (d.values())[0]
        a.append(c)
      weights_list.append(a) 
    node_weights = []
    for node in weights_list:
      node_weights.append(sum(node))
    return node_weights


REG_strength = strength(REG_weighted)
RGER_strength = strength(RGER_weighted)
RGG_strength = strength(RGG_weighted)
SF_strength = strength(SF_weighted)
SW_strength = strength(SW_weighted)

print("REG Strength")
print(REG_strength)
print()

print("RGER Strength")
print(RGER_strength)
print()

print("RGG Strength")
print(RGG_strength)
print()

print("SF Strength")
print(SF_strength)
print()

print("SW Strength")
print(SW_strength)
print()

```

##### Τοπολογία REG <b>(Regular)</b>
Περιμένουμε μέση τιμή δύναμης κοντά στο 30 καθώς τα βάρη κυμαίνονται τυχαία από 1 εώς 10 και κάθε κόμβος έχει 4 ακμές και η διασπορά θα είναι πολύ μικρή.

##### Τοπολογία RGER (<b>Random Graph Erdos-Renyi</b>)
Η κατανομή του βαθμού κόμβων ακολουθεί κατανομή Poisson άρα περιμένουμε απότομη αύξηση της γραφικής παράστασης δύναμης κόμβου και σχετικά μεγάλη διασπορά.
##### Τοπολογία RRG (<b>Random Geometric Graph</b>) 
Η κατανομή του βαθμού κόμβων ακολουθεί Ομοιόμορφη Κατανομή επομένως και στην γραφική θα δούμε ομαλή αύξηση της καμπύλης δύναμης κόμβου και η διασπορά θα είναι ακόμα πιο μεγάλη από του Random Graph. 
##### Τοπολογία SF (<b>Scale-Free Network</b>) 
H κατανομή βαθμoύ κόμβων ακολουθεί τον εκθετικό νόμο άρα στην γραφική θα δούμε μία εκτοξευση της καμπύλης δύναμης κόμβου και η διασπορά θα είναι πολύ μεγάλη σε αντιθεση με την μέση τιμή που λόγω των κόμβων με ελάχιστο βαθμό δεν θα βγει πολύ μεγάλη. 
##### Τοπολογία SW (<b>Small-World Network</b>) 
Σε αυτήν την τοπολογία έχουμε παρόμοια αποτελέσματα με του REG γιατί έχουν ίδια μέση τιμή βαθμού κόμβου. Ενώ έχουμε σημαντική διασπορά δύναμης διότι το SW degree ακολουθεί την heavy-tailed διανομή και άρα αρκετοί κόμβοι θα έχουν δύναμη που θα αποκλίνει από την μέση τιμή.


### Δ) Υπολογισμός μέσου μήκους μονοπατιού

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του στο δίκτυο. Να παρουσιαστούν συγκεντρωτικά τα αποτελέσματα. Πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως εξηγήθηκαν στη θεωρία.

```python
# χρησιμοποήστε την all_pairs_shortest_path_length του networkx 

def length_generator(G):
  pairs = nx.all_pairs_shortest_path_length(G)
  G_pairs = list (pairs) 
  length_list=[]
  for item in G_pairs:
    item_dict = item[1]
    for value in item_dict.values():
      length_list.append(value)
  return (length_list)



REG_path_lengths = length_generator(REG)
RGER_path_lengths = length_generator(RGER)
RGG_path_lengths = length_generator(RGG)
SF_path_lengths = length_generator(SF)
SW_path_lengths = length_generator(SW)


print("Mean path length REG: %f" %np.mean(REG_path_lengths))
print("Variance of path length REG: %f" %np.var(REG_path_lengths))

print("Mean path length RGER: %f" %np.mean(RGER_path_lengths))
print("Variance of path length RGER: %f" %np.var(RGER_path_lengths))

print("Mean path length RGG: %f" %np.mean(RGG_path_lengths))
print("Variance of path length RGG: %f" %np.var(RGG_path_lengths))

print("Mean path length SF: %f" %np.mean(SF_path_lengths))
print("Variance of path length SF: %f" %np.var(SF_path_lengths))

print("Mean path length SW: %f" %np.mean(SW_path_lengths))
print("Variance of path length SW: %f" %np.var(SW_path_lengths))
```

*   Οι τυχαίοι (<b>Random</b>) γράφοι έχουν average μέσο μήκος μονοπατιού.
Επειδή η μέση τιμή βαθμού κόμβων αλλά και η διασπορά του είναι μεγάλες έχει μικρό μεσο μονοπάτι. Αυτό συμβαίνει γιατί όταν ένας κόμβος έχει πολλές προσπίπτουσες ακμές είναι πιο πιθανό να συνδεθεί με τον επιθυμητό προορισμό.
*   Οι τυχαίοι γεωμετρικοί (<b>Random Geometric</b>) γράφοι έχουν μακρύ (long) μεσο μήκος μονοπατιού. Αυτό εξαρτάται πάντα από την μέγιστη ακτίνα που μπορούν να έχουν οι ακμές-όσο μεγαλύτερη ακτίνα τόσο μακρύτερο μονοπάτι. *   Τα <b>Scale-Free</b> δίκτυα έχουν μικρό (small) μέσο μήκος μονοπατιού. Oι περισσότεροι κόμβοι είτε είναι κόμβοι μεγάλου degree είτε συνδέονται με ένα hop με κόμβο μεγάλου βαθμού κόμβου άρα λαμβάνουμαι πάρα πολύ μικρή διασπορά για το μέσο μήκος μονοπατιού.

##### 2) Υπολογισμός κεντρικότητας σε μεγαλύτερες συνθετικές τοπολογίες

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να υπολογιστεί η τοπική κεντρικότητα του κάθε κόμβου, να απεικονιστεί η συγκεντρωτική κατανομή κεντρικότητας για όλους τους κόμβους και να σχολιαστούν τα αποτελέσματα για όλες τις παρακάτω μετρικές κεντρικότητας:

- <b>degree centrality: </b> ο αριθμός των ακμων που προσπίπτουν σε ένα κόμβο, διαιρεμένος δια $n-1$, όπου $n$ ο αριθμός των κόμβων.
- <b>closeness centrality: </b> καθορίζεται από την απόσταση του κόμβου από κάθε άλλο κόμβο. Όσο μικρότερη είναι αυτή η τιμή τόσο πιο κεντρικός είναι ο κόμβος.
- <b>betweenness centrality:</b> η επιρροή που έχει ένας κόμβος πάνω στη συνολική ροή. Στον πίνακα με τα συντομότερα μονοπάτια για κάθε ζεύγος κόμβων, πόσες φορές εμφανίζεται. 
- <b>eigenvector centrality:</b> εκχωρεί σχετικές βαθμολογίες σε όλους τους κόμβους του δικτύου με βάση την ιδέα ότι οι συνδέσεις με κόμβους υψηλής βαθμολογίας συμβάλλουν περισσότερο στην βαθμολογία του εν λόγω κόμβου από τις ίσες συνδέσεις με κόμβους χαμηλής βαθμολογίας.

```python
def calculate_degree_centrality(G):
   degree_centrality = nx.degree_centrality(G)
   return degree_centrality

def calculate_closeness_centrality(G):
   closeness_centrality = nx.closeness_centrality(G)
   return closeness_centrality

def calculate_betweenness_centrality(G):
   betweenness_centrality = nx.betweenness_centrality(G)
   return betweenness_centrality
REG_betweenness_centrality = calculate_betweenness_centrality(REG)

def calculate_eigenvector_centrality(G):
   eigenvector_centrality = nx.eigenvector_centrality(G)
   return eigenvector_centrality

REG_degree_centrality = calculate_degree_centrality(REG)
REG_closeness_centrality = calculate_closeness_centrality(REG)
REG_betweenness_centrality = calculate_betweenness_centrality(REG)
REG_eigenvector_centrality = calculate_eigenvector_centrality(REG)

RGER_degree_centrality = calculate_degree_centrality(RGER)
RGER_closeness_centrality = calculate_closeness_centrality(RGER)
RGER_betweenness_centrality = calculate_betweenness_centrality(RGER)
RGER_eigenvector_centrality = calculate_eigenvector_centrality(RGER)

RGG_degree_centrality = calculate_degree_centrality(RGG)
RGG_closeness_centrality = calculate_closeness_centrality(RGG)
RGG_betweenness_centrality = calculate_betweenness_centrality(RGG)
RGG_eigenvector_centrality = calculate_eigenvector_centrality(RGG)

SF_degree_centrality = calculate_degree_centrality(SF)
SF_closeness_centrality = calculate_closeness_centrality(SF)
SF_betweenness_centrality = calculate_betweenness_centrality(SF)
SF_eigenvector_centrality = calculate_eigenvector_centrality(SF)

SW_degree_centrality = calculate_degree_centrality(SW)
SW_closeness_centrality = calculate_closeness_centrality(SW)
SW_betweenness_centrality = calculate_betweenness_centrality(SW)
SW_eigenvector_centrality = calculate_eigenvector_centrality(SW)


degree_centrality = [REG_degree_centrality, RGER_degree_centrality, RGG_degree_centrality, SF_degree_centrality,
                    SW_degree_centrality]
closeness_centrality = [REG_closeness_centrality, RGER_closeness_centrality, RGG_closeness_centrality, 
                    SF_closeness_centrality, SW_closeness_centrality]
betweenness_centrality = [REG_betweenness_centrality, RGER_betweenness_centrality, RGG_betweenness_centrality, 
                    SF_betweenness_centrality, SW_betweenness_centrality]
eigenvector_centrality = [REG_eigenvector_centrality, RGER_eigenvector_centrality, RGG_eigenvector_centrality, 
                    SF_eigenvector_centrality, SW_eigenvector_centrality]

#συγκεντρωτική κατανομή των centralities στο ιδιο figure με subplots
def subplot_centrality(centrality, n, title):
    y = np.array(range(n))/float(n)
    fig, axs = plt.subplots(5)
    fig.suptitle(title)
    plt.subplots_adjust(wspace=2, hspace=2, left=0.125, right=0.9, bottom=1.1, top=2.9)
    axs[0].title.set_text("REG")
    axs[0].plot(np.sort(list(centrality[0].values())), y)
    axs[1].title.set_text("RGER")
    axs[1].plot(np.sort(list(centrality[1].values())), y)
    axs[2].title.set_text("RGG")
    axs[2].plot(np.sort(list(centrality[2].values())), y)
    axs[3].title.set_text("SF")
    axs[3].plot(np.sort(list(centrality[3].values())), y)
    axs[4].title.set_text("SW")
    axs[4].plot(np.sort(list(centrality[4].values())), y)
    
    

subplot_centrality(degree_centrality, n, "Degree Centrality")
subplot_centrality(closeness_centrality, n, "Closeness Centrality")
subplot_centrality(betweenness_centrality, n, "Betweenness Centrality")
subplot_centrality(eigenvector_centrality, n, "Eigenvector Centrality")
```

```python
print("--- Degree centrality ---")
print("Mean degree centrality REG: %f" %np.mean([x[1] for x in REG_degree_centrality.items()]))
print("Mean degree centrality RGER: %f" %np.mean([x[1] for x in RGER_degree_centrality.items()]))
print("Mean degree centrality RGG: %f" %np.mean([x[1] for x in RGG_degree_centrality.items()]))
print("Mean degree centrality SF: %f" %np.mean([x[1] for x in SF_degree_centrality.items()]))
print("Mean degree centrality SW: %f" %np.mean([x[1] for x in SW_degree_centrality.items()]))
print()

print("--- Closeness centrality ---")
print("Mean closeness centrality REG: %f" %np.mean([x[1] for x in REG_closeness_centrality.items()]))
print("Mean closeness centrality RGER: %f" %np.mean([x[1] for x in RGER_closeness_centrality.items()]))
print("Mean closeness centrality RGG: %f" %np.mean([x[1] for x in RGG_closeness_centrality.items()]))
print("Mean closeness centrality SF: %f" %np.mean([x[1] for x in SF_closeness_centrality.items()]))
print("Mean closeness centrality SW: %f" %np.mean([x[1] for x in SW_closeness_centrality.items()]))
print()

print("--- Betweenness centrality ---")
print("Mean betweenness centrality REG: %f" %np.mean([x[1] for x in REG_betweenness_centrality.items()]))
print("Mean betweenness centrality RGER: %f" %np.mean([x[1] for x in RGER_betweenness_centrality.items()]))
print("Mean betweenness centrality RGG: %f" %np.mean([x[1] for x in RGG_betweenness_centrality.items()]))
print("Mean betweenness centrality SF: %f" %np.mean([x[1] for x in SF_betweenness_centrality.items()]))
print("Mean betweenness centrality SW: %f" %np.mean([x[1] for x in SW_betweenness_centrality.items()]))
print()

print("--- Eigenvector centrality ---")
print("Mean eigenvector centrality REG: %f" %np.mean([x[1] for x in REG_eigenvector_centrality.items()]))
print("Mean eigenvector centrality RGER: %f" %np.mean([x[1] for x in RGER_eigenvector_centrality.items()]))
print("Mean eigenvector centrality RGG: %f" %np.mean([x[1] for x in RGG_eigenvector_centrality.items()]))
print("Mean eigenvector centrality SF: %f" %np.mean([x[1] for x in SF_eigenvector_centrality.items()]))
print("Mean eigenvector centrality SW: %f" %np.mean([x[1] for x in SW_eigenvector_centrality.items()]))
```

##### Σχολιάστε τα αποτελέσματα συνεκτικότητας για το REG:
Έχουμε συγκεκριμένο "βαθμό κόμβου" σε όλους τους κόμβους άρα δεν θα μπορούσε κάποιος νέος κόμβος να μην συνδεθεί με άλλους ετσι ώστε να μείνει το γράφημα μη συνεκτικό. 

```python
#για τον έλεγχο της συνεκτικότητας, χρησιμοποίηστε την is_connected του networkx
#για τη γραφική αναπαράσταση των αποτελεσμάτων, μπορείτε να χρησιμοποίησετε την bar του matplotlib
#έλεγχος συνεκτικότητας για REG

REG_generator = lambda n, d:nx.watts_strogatz_graph(n, d, 0)

def RegularGraph(n,d,fn):
  connected = 0 
  for i in range(100): 
    graph = fn(n,d)
    connected = connected + nx.is_connected(graph)
  return connected/100


#plot example:
def plotRegularGraph(n,idx,fn):
  plt.figure(idx)
  y=[RegularGraph(n,d,fn) for d in range(2,12,2)]
  plt.bar(list(range(2,12,2)),y, width=0.5, color='b')
  plt.title("REG n = %d" %n)
  plt.ylabel("Number of connected graphs")
  plt.xlabel("Degree")
  plt.show()

plotRegularGraph(100,1,REG_generator)
plotRegularGraph(200,2,REG_generator)
```

### Κ) Εγω-κεντρικότητες (Ego-centralities)
Για τις συνθετικές τοπολογίες του πίνακα 1, να υπολογιστεί η ενδιαμεσική εγω-κεντρικότητα (ego-betweenness centrality) του κάθε κόμβου. Στη συνέχεια να καταταχθούν οι κόμβοι με βάση την ενδιαμεσική εγωκεντρικότητα και την ενδιαμεσική κεντρικότητα και να συγκριθούν/σχολιαστούν οι κατατάξεις ως προς την ομοιότητά τους για κάθε μία τοπολογία.  

```python
#για την κατασκευή των εγωδικτύων, χρησιμοποιήστε την ego_graph του networkx
#για τον υπολογισμό της ενδιαμεσικής εγωκεντρικότητας θα χρειαστείτε τις adjacency_matrix (networkx), to.dense, ones (numpy)
#linalg.matrix_power, multiply, triu, nonzero του numpy
#κατασκευάστε dataframe από το pandas με γραμμές τους κόμβους και στήλες τις τιμές των υπό εξέταση κεντρικοτήτων
#κατασκευάστε στη συνέχεια pairplot από το seaborn για να εξετάσετε τις συσχετίσεις μεταξύ τους.
```

```python
d = 4
# Lattice -> p = 0
REG =  nx.watts_strogatz_graph(n, d, 0, seed=None)

M = 750
# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M)

R = 0.25
L = 1
n = 130
# για την κατασκευή του RGG, χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R, dim=2) 

d = 4
# για την κατασκευή του SF, χρησιμοποιήστε την barabasi_albert_graph από το networkx.
SF = nx.barabasi_albert_graph(n, d)

n = 130
d = 4
p = 0.3
# για την κατασκευή του SW, χρησιμοποιήστε την watts_strogatz_graph από το networkx.
SW = nx.watts_strogatz_graph(n, d, p)

```

```python
def calculate_betweenness_centrality(G):
   betweenness_centrality = nx.betweenness_centrality(G)
   return betweenness_centrality

def calculate_ego_centrality(topology):
    ego_centrality_list = []
    for node in topology.nodes():
        node_ego = nx.ego_graph(topology, node)
        # adjacency matrix
        A = (nx.adj_matrix(node_ego)).todense()
        # A^2 * (1-A)
        hadamard = np.multiply(np.matmul(A,A), (1-A))
        triu = np.triu(hadamard, 1)
        triu = triu[triu > 0]
        ego_betweenness = np.sum([1/i for i in triu])
        ego_centrality_list.append(ego_betweenness / 2)

    return ego_centrality_list   
```

```python
REG_ego = calculate_ego_centrality(REG)
RGER_ego = calculate_ego_centrality(RGER)
RGG_ego = calculate_ego_centrality(RGG)
SF_ego = calculate_ego_centrality(SF)
SW_ego = calculate_ego_centrality(SW)

REG_betweenness = calculate_betweenness_centrality(REG)
RGER_betweenness = calculate_betweenness_centrality(RGER)
RGG_betweenness = calculate_betweenness_centrality(RGG)
SF_betweenness = calculate_betweenness_centrality(SF)
SW_betweenness = calculate_betweenness_centrality(SW)
```
